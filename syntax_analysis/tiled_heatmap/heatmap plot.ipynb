```python
import pandas as pd
import numpy as np
import sys
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('LA_processed_v2.xlsx', header=1, index_col=0).fillna(0)
df = df[df.Peak != 'No Peaks found'].reset_index()
df = df.drop(columns=['index', 'Peak', 'STAT3', 'regulation'])
df2 = df.groupby('Gene').sum()
df2[df2 > 0] = 1
df2[df2 < 0] = -1
sns.clustermap(df2, cmap="coolwarm")

#sns.color_palette("hls", 8)
#fig, ax = plt.subplots(figsize=(10,10))
plt.title('LA CM1 down-regulated DE genes')
#sns.heatmap(df2, cmap="YlGnBu")
plt.savefig('LAcm1_down.pdf')
df2 = df2.replace(0,'')
df2.to_excel('LAcm1_processed_v3.xlsx')
#sns.heatmap(df2, center=0)
```
