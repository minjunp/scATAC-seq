```python
def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""
```

```python
with open('./LA_down_data_processed.csv', 'w') as f:
    f.write('Gene,Peak,')
    for i in LAcm_subs:
        if i != LAcm_subs[-1]:
            f.write(i[:-4]+',')
        else:
            f.write(i[:-4])
    f.write('\n')
    
    for gene in my_list:
        g = '/Users/minjunpark/Documents/Pitx2/manuscript/fig3/jupyter/gene_peak_intersect/LA_all/{}.bed'.format(gene)
        if os.path.getsize(g) > 0:
            # get the peaks if not empty
            df = pd.read_csv('/Users/minjunpark/Documents/Pitx2/manuscript/fig3/jupyter/gene_peak_intersect/LA_all/{}.bed'.format(gene), sep='\t', header=None)
            for i in range(len(df)):
                # write the peak
                f.write(gene + ',' + df.iloc[i,:].values[0] + ':' + str(df.iloc[i,:].values[1]) + '-' + str(df.iloc[i,:].values[2]))

                # Loop over different TFs
                for tf in LAcm_subs:
                    df_tf = pd.read_csv('/Users/minjunpark/Documents/Pitx2/manuscript/fig3/jupyter/gene_peak_intersect/LA_TF_all/{}/{}'.format(gene, tf), 
                                        header=None)
                    tf_name = find_between(tf, "", ".csv")
                    # DE direction
                    n = tf_name.lower()
                    logFC_val = LAcm.loc[n].avg_logFC
                    if df_tf.iloc[i,3] > 0: 
                        if logFC_val > 0:
                            f.write(',+1')
                        if logFC_val < 0:
                            f.write(',-1')
                    else:
                        f.write(',')                        
                f.write('\n')
        else:
            f.write(gene + ',No Peaks found')
            f.write('\n')
```
