```python
import pandas as pd
```

```python
metadata = pd.read_csv('singlecell.csv', sep=',')
```

```python
barcode = metadata[metadata.columns[0]]
```

```python
barcode
```

```python
barcodes = []
for i in barcode:
    barcodes.append(i[-1])
```

```python
barcodes
```

```python
metadata["types"] = barcodes
```

```python
metadata
```

```python
metadata.loc[metadata.types == "1", 'dtype'] = "P1_r1"
metadata.loc[metadata.types == "2", 'dtype'] = "P1_r2"
metadata.loc[metadata.types == "3", 'dtype'] = "P8_w1_r1"
metadata.loc[metadata.types == "4", 'dtype'] = "P8_w1_r2"
metadata.loc[metadata.types == "5", 'dtype'] = "P8_w2_r2"
```

```python
metadata
```

```python
metadata.to_csv('singlecell_v2.csv', sep=',', index=False)
```
