```r
library(EnhancedVolcano)
```

```r
if (!requireNamespace("BiocManager", quietly = TRUE))
    install.packages("BiocManager")

BiocManager::install("EnhancedVolcano")
```

```r
LA_ATAC_DA_peaks <- read.csv('LA_ATAC_DA_peaks.csv')
```

```r
rownames(LA_ATAC_DA_peaks) <- LA_ATAC_DA_peaks[,1]
LA_ATAC_DA_peaks[,1] <- NULL
```

```r
head(LA_ATAC_DA_peaks)
```

```r
vol <- EnhancedVolcano(LA_ATAC_DA_peaks,
    lab = rownames(LA_ATAC_DA_peaks),
    x = 'avg_logFC',
    y = 'p_val_adj',
    title = 'scATAC Left Atria Differentially Accessible peaks',
    pCutoff = 10e-3,
    FCcutoff = 0.5,
    pointSize = 3.0)
    #labSize = 5.0)
```

```r
vol
```

```r
pdf("atac_volcano.pdf", useDingbats=FALSE)
vol
dev.off()
```
