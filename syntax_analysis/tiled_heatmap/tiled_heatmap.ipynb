```r
library(readxl)
library(ggplot2) # ggplot() for plotting
library(dplyr) # data reformatting
library(tidyr) # data reformatting
library(stringr) # string manipulation
library(reshape2)
library(RColorBrewer)
library(repr)
```

```r
LAcm1_path <- '/Users/minjunpark/Documents/syntax_analysis/data/LAcm1_processed_v3.xlsx'
PVcm1_path <- '/Users/minjunpark/Documents/syntax_analysis/data/PVcm1_processed_v3.xlsx'
PVcm2_path <- '/Users/minjunpark/Documents/syntax_analysis/data/PVcm2_processed_v3.xlsx'
```

```r
options(repr.plot.width=7, repr.plot.height=7)

d <- read_excel(LAcm1_path)
# fillna with 0
d[is.na(d)] <- 0
df <- as.data.frame(d)
rownames(df) <- df$Gene
df$Gene <- NULL
df <- as.matrix(df)
#print(colnames(df))
#print(rownames(df))
# Melt it to single vectors
df_melt <-melt(df)
df_melt$value <- as.factor(df_melt$value)

p <- ggplot(df_melt, aes(x = Var2, y = Var1, fill=value)) + 
  geom_tile(colour="white",size=0.3)+
  coord_equal() + 
  theme(axis.text.x = element_text(angle = 60, hjust = 1)) +
  guides(fill=guide_legend(title="value"))+
  labs(x="TFs",y="DE Genes",title="LA CM1") +
  scale_fill_manual(breaks = c(1, 0, -1), values=c("#99d594","#ffffbf","#fc8d59")) 
p
#export figure
ggsave(p,filename="LA_cm1.pdf")
```
