```python
import pandas as pd
import numpy as np
import sys
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
```

# LA CM1

```python
df = pd.read_excel('LAcm1_processed_v3.xlsx', header=0, index_col=0).fillna(0)
```

```python
xlabels = df.columns.values
M = len(xlabels)
ylabels = df.index.values
N = len(ylabels)

x, y = np.meshgrid(np.arange(M), np.arange(N))

s = np.ones(shape=(N,M)) # Size of the bubble
c = np.array(df)

fig, ax = plt.subplots(figsize=(7,30))

R = s/3
circles = [plt.Circle((j,i), radius=r) for r, j, i in zip(R.flat, x.flat, y.flat)]
col = PatchCollection(circles, array=c.flatten(), cmap="seismic")
ax.add_collection(col)

ax.set(xticks=np.arange(M), yticks=np.arange(N),
       xticklabels=xlabels, yticklabels=ylabels)
ax.set_xticks(np.arange(M+1)-0.5, minor=True)
ax.set_yticks(np.arange(N+1)-0.5, minor=True)
ax.grid(which='minor')

#fig.colorbar(col)
plt.title('LA CM1 Circle Plot')
plt.savefig('LAcm1_circleplot.pdf')
plt.show()
```
